## Data Preperation and cleaning

- Load the file using pandas
- Look at some information about the dataand the columns
- Fix any missing or incorrect values

```python
import pandas as pd
```

```python
df = pd.read_csv(data_filename)
```

```python
df.columns
```

```python
df.info()
```

```python
df.describe()
```

```python
# To determine how many numeric columns are present
numerics = ['int16','int32','int64','float16','float32','float64']
numeric_df = df.select_dtypes(include=numerics)
len(numeric_df.columns)
```

Percentage of missing values per column

```python
missing_percentage = df.isna().sum().sort_values(ascending = False) / len(df)
missing_percentage
# In this the isna() function converts the data into bool.
# If the value is missing it marks it as true
#isna().sum gives us all the missing values in all columns
```

```python
missing_percentage = missing_percentage[missing_percentage != 0]
# This removes all the values which are zero
```

```python
missing_percentage.plot(kind='barh')
```

### City

```python
cities = df.City.unique()
len(cities)
```

```python
cities_by_accident = df.City.value_counts()
cities_by_accident
# Here we can sort the cities with highest number of accidents
```

```python
cities_by_accident[:10]
# Top 10 cities_by_Accident
```

```python
'New York' in df.City
# New York not present in the data
```

```python
cities_by_accident[:20].plot(kind='barh')
```

```python
import seaborn as sns
sns.set_style("darkgrid")
```

```python
sns.histplot(cities_by_accident, log_scale = True)
# This gives a better picture of how many accidents happen 
```

```python
high_accident_cities = cities_by_accident[cities_by_accident > 1000]
low_accident_cities = cities_by_accident[cities_by_accident <= 1000]
```

```python
len(high_accident_cities) / len(cities)
# Less than 5% of cities have accidents more than 1000
```

```python
cities_by_accident[cities_by_accident == 1]
# Over 1110 cities have reported only 1 accident
```

### Start Time

```python
df.Start_Time
```

```python
df.Start_Time = pd.to_datetime(df.Start_Time)
# Convert the string into date time format
```

```python
sns.histplot(df.Start_Time.dt.hour, bins = 24)
# Converting the timestamp into hour and plotting the histogram based on hour
```

```python
sns.distplot(df.Start_Time.dt.hour, bins=24, kde = False, norm_hist= True)
```

- A high percentage of accidents occur between 12PM to 5 PM
- Next highest is from 7AM - 10 AM

```python
sns.distplot(df.Start_Time.dt.dayofweek, bins=7, kde = False, norm_hist= True)
```

- Accidents are less on weekends.

```python
#  5 - saturday , 6 - Sunday
Sunday_Start_Time = df.Start_Time[df.Start_Time.dt.dayofweek == 6]
sns.distplot(Sunday_Start_Time.dt.hour, bins=24, kde = False, norm_hist= True)
```

```python
Monday_Start_Time = df.Start_Time[df.Start_Time.dt.dayofweek == 0]
sns.distplot(Monday_Start_Time.dt.hour, bins=24, kde = False, norm_hist= True)
```

Is the distribution of accidents by hour the same on weekends as on weekdays.
- Yes the trend on sunday and monday are different even tho the peak accidents happen during 2-5 PM

```python
sns.distplot(df.Start_Time.dt.month, bins=12, kde = False, norm_hist= True)
```

```python
df_2017 = df[df.Start_Time.dt.year == 2017]
sns.distplot(df_2017.Start_Time.dt.month, bins=12, kde = False, norm_hist= True)
```

```python
df_2018 = df[df.Start_Time.dt.year == 2018]
sns.distplot(df_2018.Start_Time.dt.month, bins=12, kde = False, norm_hist= True)
```

-when we check the yearwise data ,the data from2017 and 2018 are more balanced compared to other years.
-This is because there is more inconsistent data in those year.
